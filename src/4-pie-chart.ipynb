### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.  
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = x ** 2  # Compute y as the square of x

# Create the line plot
plt.plot(x, y, label='y = x^2', color='b', linewidth=2)

# Customize the line plot
plt.title('Square Function Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the line plot
plt.show()

# Create bar plot data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Create the bar plot
plt.bar(categories, values, color=colors)

# Customize the bar plot
plt.title('Category Value Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the bar plot
plt.show()

# Create scatter plot data
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.scatter(x_scatter, y_scatter, color='red')

# Annotate each point
for i in range(len(x_scatter)):
    plt.annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='center')

# Customize the scatter plot
plt.title('Scatter Plot with Annotations')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True, linestyle='--', alpha=0.6)

# Show the scatter plot
plt.show()

# Create pie chart data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'yellow'], startangle=140)

# Customize the pie chart
plt.title('Programming Language Usage')

# Show the pie chart
plt.show()
```
