### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = x ** 2  # Compute y as the square of x

# Line plot in first subplot
axs[0, 0].plot(x, y, label='y = x^2', color='b', linewidth=2)
axs[0, 0].set_title('Square Function Plot')
axs[0, 0].set_xlabel('X Values')
axs[0, 0].set_ylabel('Y Values')
axs[0, 0].grid(True, linestyle='--', alpha=0.6)
axs[0, 0].legend()

# Bar plot in second subplot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title('Category Value Bar Plot')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot in third subplot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='red')
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='center')
axs[1, 0].set_title('Scatter Plot with Annotations')
axs[1, 0].set_xlabel('X Values')
axs[1, 0].set_ylabel('Y Values')
axs[1, 0].grid(True, linestyle='--', alpha=0.6)

# Pie chart in fourth subplot
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'yellow'], startangle=140)
axs[1, 1].set_title('Programming Language Usage')

# Adjust layout
plt.tight_layout()
plt.show()
```
